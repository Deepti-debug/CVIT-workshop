### Problem 1

Given a list of binary strings `numbers`. Define a function that takes the first `k` strings and concats their parity into a string `encoding` and returns it.

````python
'''
By me
'''

def str_encoding(numbers, k):
  encoding = ""
  for index in range(k):
    parity = 0
    for each_binary_digit in numbers[index]:
      parity = parity ^ int(each_binary_digit)
    print(f"For \"k\" = {numbers[index]}, parity is {parity}")
    print("----")
    encoding = encoding + str(parity) 
  return encoding

numbers = ["000", "100", "010", "110", "001", "101", "011", "111"]

while(1):
  k = input("Input an integer corresponding to how many elements of the list you want to consider for the task: ")
  try:
    int_k = int(k)
    break
  except:
    print("Not an Integer !!")
    continue

output = str_encoding(numbers, int_k)
print(f"Final encoded string is {output}")
````

**Solution**:

````python
#@title
def get_encoding(numbers,k):
  encoding = ""
  for index in range(k):
    parity = 0
    for digit in numbers[index]:
        parity ^= int(digit)
    encoding += str(parity)
  return encoding
````
